```python
#Paquetes necesarios para el programa
import numpy as np
from numpy import cos, pi
from numpy.random import uniform
#Pedimos x, iteraciones y mutación al usuario
d=int(input("Introduzca el máximo valor de x (no mayor que 7): "))
I=int(input("Introduzca el número de iteraciones: "))
m=float(input("Introduzca la probabilidad de mutación: "))
#Declaramos los valores del dominio, el tamaño de la población, el # de iteraciones y la mutación
Dominio=np.arange(d+1)
Tamaño=3
Iteraciones=I
Mutación=m
#Definimos algunas funciones que vamos a usar
def F(x): #La función que vamos a maximizar
    return x*cos(pi*x)
def Aptitud(x): #Devuelve el valor de x con mejor aptitud en una población
    A=x[0]
    for i in x:
        if F(A)<F(i):
            A=i
    return A
def Seleccion(x): #Devuelve los dos padres más aptos a seleccionar de una población
    P1=Aptitud(x)
    if P1==x[0]:
        if F(x[1])<F(x[2]):
            P2=x[2]
        else:
            P2=x[1]
    if P1==x[1]:
        if F(x[0])<F(x[2]):
            P2=x[2]
        else:
            P2=x[0]
    if P1==x[2]:
        if F(x[0])<F(x[1]):
            P2=x[1]
        else:
            P2=x[0]
    return np.array([P1,P2])
def binario(x): #Convierte los valores de x a binario
    if x==0:
        return np.array([0,0,0])
    if x==1:
        return np.array([0,0,1])
    if x==2:
        return np.array([0,1,0])
    if x==3:
        return np.array([0,1,1])
    if x==4:
        return np.array([1,0,0])
    if x==5:
        return np.array([1,0,1])
    if x==6:
        return np.array([1,1,0])
    if x==7:
        return np.array([1,1,1])
def Binario(x): #Escribe en binario los valores de los padres seleccionados de una población
    return np.array([binario(x[0]),binario(x[1])])
def Cruce(x): #Cruza los dos padres de acuerdo a un cruce en múltiples puntos
    Aleatorio=np.random.randint(0,3)
    h1=x[0]
    h2=x[1]
    H1=np.copy(x[0])
    H2=np.copy(x[1])
    H1[Aleatorio]=h2[Aleatorio]
    H2[Aleatorio]=h1[Aleatorio]
    return np.array([H1,H2])
def Mutacion(x): #Muta a los hijos con base en la probabilidad de mutación
    M1=x[0]
    M2=x[1]
    A1=uniform(size=3)
    A2=uniform(size=3)
    for i in range(3):
        if A1[i]<Mutación:
            M1[i]=1-M1[i]
        if A2[i]<Mutación:
            M2[i]=1-M2[i]
    return np.array([M1,M2])
def Conversion(x): #Reescribe los hijos como números naturales
    C1=x[0]
    C2=x[1]
    c1=C1[0]*4 + C1[1]*2 + C1[2]*1
    c2=C2[0]*4 + C2[1]*2 + C2[2]*1
    return np.array([c1,c2])
def Factibilidad(x): #Analiza la factibilidad de los hijos 
    f1=0
    f2=0
    if x[0] in Dominio:
        f1=0
    else:
        f1=1
    if x[1] in Dominio:
        f2=0
    else:
        f2=1
    f=f1+f2
    return f
def Reemplazo(x,y): #Reemplaza los valores menos aptos de la población con los hijos
    i=0
    j=0
    C1=0
    C2=0
    E1=0
    E2=0
    if F(x[0])<F(x[1]):
        i=0
    else:
        i=1
    if F(x[1-i])<F(x[2]):
        j=1-i
    else:
        j=2
    if F(x[i])<F(x[j]):
        C1=np.copy(i)
        C2=np.copy(j)
    else:
        C1=np.copy(j)
        C2=np.copy(i)
    if F(y[0])<F(y[1]):
        E1=y[1]
        E2=y[0]
    else:
        E1=y[0]
        E2=y[1]
    if E1 in x and E2 not in x:
        if F(x[C1])<F(E2):
            x[C1]=E2
            x[C2]=x[C2]
        else:
            x[C1]=x[C1]
            if F(x[C2])<F(E2):
                x[C2]=E2
            else:
                x[C2]=x[C2]  
    if E1 not in x and E2 in x:
        if F(x[C1])<F(E1):
            x[C1]=E1
            x[C2]=x[C2]
        else:
            x[C1]=x[C1]
            if F(x[C2])<F(E1):
                x[C2]=E1
            else:
                x[C2]=x[C2]
    if E1 not in x and E2 not in x:
        if F(x[C1])<F(E1):
            x[C1]=E1
            if F(x[C2])<F(E2):
                 x[C2]=E2
            else:
                 x[C2]=x[C2]
        else:
            x[C1]=x[C1]
            if F(x[C2])<F(E1):
                x[C2]=E1
            else:     
                x[C2]=x[C2]
    if E1 in x and E2 in x:
        x[C1]=x[C1]
        x[C2]=x[C2]
    return x
print("\n-------------------------")
print("-------------------------")
print("\nDominio factible: x en %s"%(Dominio)) #Muestra el dominio factible
print("\nTamaño de la población: %s"%(Tamaño)) #Muestra el tamaño de la población
print("\nIteraciones: %s"%(Iteraciones)) #Muestra el número de iteraciones
print("\nProbabilidad de mutación: %s"%(Mutación)) #Muestra la probabilidad de mutación
#Generamos la primera población aleatoriamente
Poblacion=np.array([0,0,0])
Poblacion[0]=np.random.randint(0,6)
for i in range(1,Tamaño):
    j=0
    while j==0:
        n=np.random.randint(0,6)
        if n in Poblacion:
            j=0
        else:
            Poblacion[i]=n
            j=1
print("\n-------------------------")
print("-------------------------")
print("\nIteración #%s"%(0)) #Muestra la iteración actual
print("\nPoblación: %s"%(Poblacion)) #Muestra la población actual
print("\nMejor solución: x=%s"%(Aptitud(Poblacion))) #Muestra la mejor solución hasta el momento
#Iniciamos las iteraciones
for i in range(1,Iteraciones+1):
    print("\n--------------------------")
    print("\nIteración #%s"%(i)) #Muestra la iteración actual
    print("\nPoblación: %s"%(Poblacion)) #Muestra la población inicial
    #Ciclo de factibilidad de las soluciones
    f=1
    while f!=0:
        S=Seleccion(Poblacion) #Seleccionamos los padres más aptos de la población
        #print("\n[P1 P2]=%s"%(S))
        B=Binario(S) #Convertimos a binario los padres
        #print("\n[P1 P2]_Binario=%s"%(B))
        C=Cruce(B) #Cruzamos los padres
        #print("\n[H1 H2]_Binario=%s"%(C))
        M=Mutacion(C) #Mutamos los hijos
        #print("\n[H1 H2]_BinarioM=%s"%(M))
        c=Conversion(M) #Convertimos los hijos a naturales
        print("\n[H1 H2]=%s"%(c))
        f=Factibilidad(c) #Analizamos si pertenecen al dominio 
    Poblacion=Reemplazo(Poblacion,c) #Actualizamos la población con los hijos
    print("\nNueva población: %s"%(Poblacion)) #Muestra la nueva población
    print("\nMejor solución: x=%s"%(Aptitud(Poblacion))) #Muestra la mejor solución actual
    if i==Iteraciones:
        print("\nSe llegó al máximo de iteraciones propuestas.")
```
