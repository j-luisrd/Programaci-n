# **Cuál ha sido el país más ganador en los olímpicos?**

```python
import numpy as np
import pandas as pd
```

```python
tab=pd.read_csv('olympics.csv', index_col = 0, skiprows=1)
tab.head()
```

```python
k=0
for col in tab.columns:
    if col[:2]=='01':
        if 1<=k<=3:
            tab.rename(columns={col:'Gold Summer'}, inplace=True)
        if 6<=k<=8:
            tab.rename(columns={col:'Gold Winter'}, inplace=True)
        else: 
            tab.rename(columns={col:'Gold Total'}, inplace=True)
    if col[:2]=='02':
        if 1<=k<=3:
            tab.rename(columns={col:'Silver Summer'}, inplace=True)
        if 6<=k<=8:
            tab.rename(columns={col:'Silver Winter'}, inplace=True)
        else: 
            tab.rename(columns={col:'Silver Total'}, inplace=True)
    if col[:2]=='03':
        if 1<=k<=3:
            tab.rename(columns={col:'Bronze Summer'}, inplace=True)
        if 6<=k<=8:
            tab.rename(columns={col:'Bronze Winter'}, inplace=True)
        else: 
            tab.rename(columns={col:'Bronze Total'}, inplace=True)
    if col[:1]=='№':
        if k==0:
            tab.rename(columns={col:'Summer Games'}, inplace=True)
        if k==5:
            tab.rename(columns={col:'Winter Games'}, inplace=True)
        else: 
            tab.rename(columns={col:'Total Games'}, inplace=True) 
    if col[:1]=='T':
        if k==4:
            tab.rename(columns={col:'Summer Total'}, inplace=True)
        else:
            tab.rename(columns={col:'Winter Total'}, inplace=True)
    k=k+1
tab.head()
```
