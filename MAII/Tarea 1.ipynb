```python
import numpy as np
import matplotlib.pyplot as plt
```

### **Apply the summation technique of this section to show that ** $\displaystyle \sum_{n=1}^\infty \frac{\sin(nx)}{n}=\left\{\begin{aligned} \frac{1}{2}(\pi-x)&, \ &0<x\leq \pi,\\ -\frac{1}{2}(\pi+x)&, \ &-\pi\leq x<0. \end{aligned}\right.$

```python
def f(x,n):
    s=0
    for i in range(1,n+1):
        s+=np.sin(i*x)/i
    return s

def g(x):
    return (np.pi-x)/2

def h(x):
    return -(np.pi+x)/2

X1=np.linspace(0,np.pi,50)
X2=np.linspace(-np.pi,0,50)

F1=np.zeros(len(X1))
F2=np.zeros(len(X1))
G=np.zeros(len(X1))
H=np.zeros(len(X1))

for i in range(len(X1)):
    F1[i]=f(X1[i],30)
    F2[i]=f(X2[i],30)
    G[i]=g(X1[i])
    H[i]=h(X2[i])

plt.figure(figsize=(10,6), dpi=80)
ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))
plt.plot(X1, F1, color="red", linewidth=2.5, linestyle="-", label="r'$x\cos(\pi x)$'")
plt.plot(X2, F2, color="red", linewidth=2.5, linestyle="-", label="r'$x\cos(\pi x)$'")
           
plt.plot(X1, G, color="blue", linewidth=2.5, linestyle="-", label="r'$x\cos(\pi x)$'")
plt.plot(X2, H, color="blue", linewidth=2.5, linestyle="-", label="r'$x\cos(\pi x)$'")

for label in ax.get_xticklabels() + ax.get_yticklabels():
    label.set_fontsize(16)
    label.set_bbox(dict(facecolor='white', edgecolor='None', alpha=0.65 ))
plt.show()
```

In the graph, the red line is the plot of the given series and the blue line is the plot of the given function. The plot of the series is for the first 30 terms; if we add more terms to plot the series, the red line will be over the blue one, verifying in a graphical way, the equality of the function and the series.

### **Sum the series** $\displaystyle \sum_{n=1}^\infty \frac{(-1)^{n+1}\sin(nx)}{n}$ **and show that it equals** $\displaystyle \frac{x}{2}$.

```python
def f(x,n):
    s=0
    for i in range(1,n+1):
        s+=(((-1)**(i+1))*np.sin(i*x))/i
    return s

def g(x):
    return x/2

X=np.linspace(-10,10,100)

F=np.zeros(len(X))
G=np.zeros(len(X))

for i in range(len(X)):
    F[i]=f(X[i],50)
    G[i]=g(X[i])

plt.figure(figsize=(10,6), dpi=80)
ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))

plt.plot(X, F, color="red", linewidth=2.5, linestyle="-", label="r'$x\cos(\pi x)$'")
           
plt.plot(X, G, color="blue", linewidth=2.5, linestyle="-", label="r'$x\cos(\pi x)$'")

for label in ax.get_xticklabels() + ax.get_yticklabels():
    label.set_fontsize(16)
    label.set_bbox(dict(facecolor='white', edgecolor='None', alpha=0.65 ))
plt.show()
```

In the graph, the red line is the plot of the given series and the blue line is the plot of the given function. The plot of the series is for the first 50 terms; if we join the parts of the red line that are in a "diagonal" way we will get the blue line, verifying in a graphical way, the equality of the function and the series.

### **Sum the trigonometric series** $\displaystyle \sum_{n=0}^\infty\frac{\sin((2n+1)x)}{2n+1}$ **and show that it equals** $$\left\{\begin{aligned} \frac{\pi}{4}&, \ &0<x<\pi,\\ -\frac{\pi}{4}&, \ &-\pi<x<0. \end{aligned}\right.$$

```python
def f(x,n):
    s=0
    for i in range(n+1):
        s+=np.sin((2*i*x)+x)/(2*i+1)
    return s

def g(x):
    return np.pi/4

def h(x):
    return -np.pi/4

X1=np.linspace(0,np.pi,100)
X2=np.linspace(-np.pi,0,100)

F1=np.zeros(len(X1))
F2=np.zeros(len(X1))
G=np.zeros(len(X1))
H=np.zeros(len(X1))

for i in range(len(X1)):
    F1[i]=f(X1[i],50)
    F2[i]=f(X2[i],50)
    G[i]=g(X1[i])
    H[i]=h(X2[i])

plt.figure(figsize=(10,6), dpi=80)
ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))
plt.plot(X1, F1, color="red", linewidth=2.5, linestyle="-", label="r'$x\cos(\pi x)$'")
plt.plot(X2, F2, color="red", linewidth=2.5, linestyle="-", label="r'$x\cos(\pi x)$'")
           
plt.plot(X1, G, color="blue", linewidth=2.5, linestyle="-", label="r'$x\cos(\pi x)$'")
plt.plot(X2, H, color="blue", linewidth=2.5, linestyle="-", label="r'$x\cos(\pi x)$'")

for label in ax.get_xticklabels() + ax.get_yticklabels():
    label.set_fontsize(16)
    label.set_bbox(dict(facecolor='white', edgecolor='None', alpha=0.65 ))
plt.show()
```

In the graph, the red line is the plot of the given series and the blue line is the plot of the given function. The plot of the series is for the first 50 terms; if we add more terms to plot the series, the red line will be over the blue one, verifying in a graphical way, the equality of the function and the series.
